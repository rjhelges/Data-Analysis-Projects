```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
movies = pd.read_csv("fandango_score_comparison.csv")

movies
```

```python
plt.hist(movies['Metacritic_norm_round'])
```

```python
plt.hist(movies['Fandango_Stars'], 6)
```

Metacritic has a wider range of ratings, from 0.5 to 4.5. Fandango's range is much less, from 3.0 to 5.0. The mode for Metacritic is at 3.0 while for Fandango it's at 4.5.

```python
print(movies['Fandango_Stars'].mean())
print(movies['Metacritic_norm_round'].mean())
```

```python
print(movies['Fandango_Stars'].median())
print(movies['Metacritic_norm_round'].median())
```

```python
print(movies['Fandango_Stars'].std())
print(movies['Metacritic_norm_round'].std())
```

So both mean and median are higher for Fandango, confirming what the histogram showed us. Also the standard deviation for Fandago is small than Metacritic, because the ratings range is smaller.

```python
plt.scatter(movies['Metacritic_norm_round'], movies['Fandango_Stars'])
```

```python
movies['fm_diff'] = movies['Fandango_Stars'] - movies['Metacritic_norm_round']
movies['fm_diff'] = movies['fm_diff'].abs()
```

```python
movies.sort_values('fm_diff', ascending=False, inplace=True)
movies.head()
```

```python
import scipy.stats as spy

r_value, p_value = spy.pearsonr(movies['Fandango_Stars'], movies['Metacritic_norm_round'])
r_value
```

The correlation between the 2 scores is fairly low, which matches the spread out nature of the scatterplot from earlier. So how one site rates a movie does not appear to determine how the other with rate the same movie.

```python
slope, intercept, r_value, p_value, std_err = spy.linregress(movies['Metacritic_norm_round'], movies['Fandango_Stars'])
```
