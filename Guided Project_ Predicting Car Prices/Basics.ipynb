```python
import pandas as pd
import numpy as np

pd.options.display.max_columns = 99
```

```python
cols = ['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 
        'drive-wheels', 'engine-location', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-type', 
        'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
cars = pd.read_csv('imports-85.data', names=cols)
```

```python
cars = cars.replace('?', np.nan)
```

```python
continuous_values_cols = ['normalized-losses', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
numeric_cars = cars[continuous_values_cols]
```

```python
numeric_cars
```

```python
numeric_cars = numeric_cars.astype('float')
numeric_cars.isnull().sum()
```

```python
numeric_cars = numeric_cars.drop('normalized-losses', axis=1)
```

```python
numeric_cars
```

```python
numeric_cars = numeric_cars.dropna()
```

```python
numeric_cars.isnull().sum()
```

```python
price_col = numeric_cars['price']
numeric_cars = (numeric_cars - numeric_cars.min())/(numeric_cars.max() - numeric_cars.min())
numeric_cars['price'] = price_col
```

```python
numeric_cars
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

def knn_train_test(features, target, df):
    new_index = np.random.permutation(df.index)
    new_df = df.reindex(new_index)
    
    split_index = int(len(df) / 2)
    train = df[0:split_index]
    test = df[split_index:]
    
    knn = KNeighborsRegressor()
    knn.fit(train[features], train[target])
    prediction = knn.predict(test[features])
    mse = mean_squared_error(prediction, test[target])
    rsme = mse ** (1/2)
    return rsme
```

```python
for i in numeric_cars.columns:
    print(i + ": " + str(knn_train_test([i], ['price'], numeric_cars)))
```

City MPG, horsepower, and highway MPG had the 3 lowest RMSE of the univariate models with the default K-nearest neighbors settings.

```python
def knn_train_test_k(features, target, df, k):
    new_index = np.random.permutation(df.index)
    new_df = df.reindex(new_index)
    
    split_index = int(len(df) / 2)
    train = df[0:split_index]
    test = df[split_index:]
    
    knn = KNeighborsRegressor(n_neighbors=k)
    knn.fit(train[features], train[target])
    prediction = knn.predict(test[features])
    mse = mean_squared_error(prediction, test[target])
    rsme = mse ** (1/2)
    return rsme
```

```python
import collections

k_values = [1, 3, 5, 7, 9]
col_rsme = {}
cols = numeric_cars.drop('price', axis=1)
for i in cols:
    k_dict = collections.OrderedDict()
    for k in k_values:
        rmse = knn_train_test_k([i], ['price'], numeric_cars, k)
        k_dict[k] = rmse
    col_rsme[i] = k_dict
```

```python
col_rsme
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

for k,v in col_rsme.items():
    x = list(v.keys())
    y = list(v.values())
    
    plt.plot(x,y)
    plt.xlabel('k value')
    plt.ylabel('RMSE')
```

```python
three_features = ['city-mpg', 'horsepower', 'highway-mpg']
k_values = [i for i in range(1, 26)]
cols = numeric_cars.drop('price', axis=1)
three_rsme = collections.OrderedDict()
for k in k_values:
    rmse = knn_train_test_k(three_features, ['price'], numeric_cars, k)
    three_rsme[k] = rmse

four_features = ['city-mpg', 'horsepower', 'highway-mpg', 'curb-weight']
four_rsme = collections.OrderedDict()
for k in k_values:
    rmse = knn_train_test_k(four_features, ['price'], numeric_cars, k)
    four_rsme[k] = rmse

two_features = ['city-mpg', 'horsepower']
two_rsme = collections.OrderedDict()
for k in k_values:
    rmse = knn_train_test_k(two_features, ['price'], numeric_cars, k)
    two_rsme[k] = rmse
    
five_features = ['city-mpg', 'horsepower', 'highway-mpg', 'curb-weight', 'width']
five_rsme = collections.OrderedDict()
for k in k_values:
    rmse = knn_train_test_k(five_features, ['price'], numeric_cars, k)
    five_rsme[k] = rmse
    
rsme_values = {}
rsme_values['three-features'] = three_rsme
rsme_values['four-features'] = four_rsme
rsme_values['two-features'] = two_rsme
rsme_values['five-features'] = five_rsme

rsme_values
```
