```python
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
```

```python
star_wars.head(10)
```

```python
star_wars.columns
```

```python
respondent_filter = star_wars["RespondentID"].notnull()
star_wars = star_wars[respondent_filter]
```

```python
star_wars.head()
```

```python
yes_no = {
    "Yes": True,
    "No": False
}

star_wars["Have you seen any of the 6 films in the Star Wars franchise?"] = star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].map(yes_no)
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] = star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].map(yes_no)
```

```python
star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].value_counts()
```

```python
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].value_counts()
```

```python
import numpy as np

movie_titles = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}

for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(movie_titles)
```

```python
star_wars=star_wars.rename(columns={
    "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
    "Unnamed: 4": "seen_2",
    "Unnamed: 5": "seen_3",
    "Unnamed: 6": "seen_4",
    "Unnamed: 7": "seen_5",
    "Unnamed: 8": "seen_6",
})
```

```python
star_wars.head()
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
star_wars=star_wars.rename(columns={
    "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
    "Unnamed: 10": "ranking_2",
    "Unnamed: 11": "ranking_3",
    "Unnamed: 12": "ranking_4",
    "Unnamed: 13": "ranking_5",
    "Unnamed: 14": "ranking_6",
})
```

```python
star_wars.head()
```

```python
rankings = star_wars[star_wars.columns[9:15]].mean()
```

```python
%matplotlib inline

rankings = star_wars[star_wars.columns[9:15]].mean()
```

So after cleaning up the data by changing yes/no values to True/False, renaming columns, and changing the ranking type to float, we are able to look at a chart of the rankings of each Star Wars movie. It shows that episodes IV, V, and VI at more popular then episodes I, II, and III with episode V being the most popular and episode III being the least. This is not surprising because the original Star Wars trilogy (IV, V, and VI) has always been believed to be more well liked the the prequels (I, II, III).

```python
movies_seen = star_wars[star_wars.columns[3:9]].mean()
```

```python
movies_seen.plot.bar()
```

Here you see episodes V, VI, and I were seen the most. This makes sense as the popularity in Star Wars rose after episode IV when it was a relatively unknown movie. Also with V and VI being the 2 most popular movies, it's no surprised they are also the 2 most watched. I suspect episode I is so high because of the anticipation of a new Star Wars movie for the 1st time 15 or so years. But the lower quality of episode 1 cause the rest of the prequels, II and III, to suffer.
